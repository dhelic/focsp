## Insertion sort

Insertion sort has a linear run time in the best case and a quadratic run time in the worst case.

```python
import random
import timeit
import matplotlib.pyplot as plt


def insertion_sort(l):
    if len(l) < 2:
        return
    
    for i in range(1, len(l)):
        current_value = l[i]
        j = i - 1
        while j >= 0 and current_value < l[j]:
            l[j + 1] = l[j]
            j -= 1
        l[j + 1] = current_value
```

### Testing insertion sort

```python
def test():
    l = [4, 3, 1, 2, 6]
    insertion_sort(l)
    print(l)
    

test()
```

### Measuring time for insertion sort in the best and worst case

```python
n_max = 1000
avg_sorted = []
avg_reversed = []

for n in range(100, n_max, 100):
    print(n)
    l = random.sample(range(1, 100000), n)
    l_sorted = l.copy()
    l_sorted.sort()
    l_reversed = l.copy()
    l_reversed.sort(reverse=True)

    result_sorted = []
    result_reversed = []

    for i in range(100):
        l1 = l_sorted.copy()
        result_sorted.append(timeit.timeit('insertion_sort(l1)', globals=globals(), number=1))
        l1 = l_reversed.copy()
        result_reversed.append(timeit.timeit('insertion_sort(l1)', globals=globals(), number=1))

    avg_sorted.append(1000 * (sum(result_sorted) / 100))
    avg_reversed.append(1000 * (sum(result_reversed) / 100))

plt.plot(range(100, n_max, 100), avg_sorted, label='Best case')
plt.plot(range(100, n_max, 100), avg_reversed, label='Worst case')
plt.title('Insertion sort run time')
plt.xlabel('Input size $n$')
plt.ylabel('Time in ms')
plt.legend()
```

## Growth of functions

Show that $f(x) = x^2 + 2x + 1$ is $O(x^2)$.

```python
import numpy as np


def f_x(x):
    return x ** 2 + 2 * x + 1


def g_x(x):
    return x ** 2


x = np.linspace(0, 2, 101)
f = f_x(x)
g = g_x(x)
o_g = 4 * g

plt.figure(1)
plt.plot(x, f, label='$x^2 + 2x + 1$')
plt.plot(x, g, label='$x^2$')
plt.plot(x, o_g, label='$4x^2$')
plt.title('$x^2 + 2x + 1$ is $O(x^2)$')
plt.xlabel('$x$')
plt.ylabel('$f(x), g(x), O(g(x))$')
plt.ylim([0, 6])
plt.legend()

o_g = 3 * g

plt.figure(2)
plt.plot(x, f, label='$x^2 + 2x + 1$')
plt.plot(x, g, label='$x^2$')
plt.plot(x, o_g, label='$3x^2$')
plt.title('$x^2 + 2x + 1$ is $O(x^2)$')
plt.xlabel('$x$')
plt.ylabel('$f(x), g(x), O(g(x))$')
plt.ylim([0, 8])
plt.legend()
```
